1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Init parameters
N = 1200
mu = 0

# Set random seed for consistency
np.random.seed(42)

# Generate the variables
x_1 = np.random.normal(mu, scale=1, size=N)
x_2 = x_1 + np.random.normal(mu, scale=3, size=N)
x_3 = 2*x_1 + x_2
x_arr = np.stack([x_1, x_2, x_3])

# Get covariance matrix
cov_matrix = np.cov(x_arr)

# Get eigenvectors and eigenvalues of the cov matrix
cov_eigen = np.linalg.eig(cov_matrix)

# Now get eigenvecs and eigenvals with SVD
svd_eigen = np.linalg.svd(x_arr)

# Check that they are the same
print('Eigenvalues:')
print('SVD:\n',np.round(np.sort(svd_eigen[1]**2/N), 4))
print('Covariance matrix eigenvalues:\n',np.round(np.sort(cov_eigen[0]),4))

print('\nEigenvectors (not sorted):')
print('SVD:\n', np.round(svd_eigen[0],4))
print('Covariance matrix eigenvectors:\n', np.round(cov_eigen[1],4))
```

```python
print('Variability explained by all PCs:',\
      np.round(np.sum(cov_eigen[0][0:3]) / np.trace(np.cov(x_arr)), 4))
```

```python
print('Variability explained with PCs 1 and 3:',\
     np.round(np.sum(cov_eigen[0][[0,2]]) / np.trace(np.cov(x_arr)), 4))
```

```python
# Let's redefine the data in the PCA basis
V0 = cov_eigen[1].copy()
x_arr_t = np.dot(V0.T, x_arr)
```

```python
# Plot the og data, and the transformed data
FONTSIZE = 20
fig, axs = plt.subplots(figsize=(13,8), dpi=200,
                       nrows=2, ncols=3)
for i, comb in enumerate([[0,1],[1,2], [0,2]]):
    axs[0,i].scatter(x_arr[comb[0],:], x_arr[comb[1],:])
    axs[1,i].scatter(x_arr_t[comb[0],:], x_arr_t[comb[1],:])
    # Set xlabels
    axs[0,i].set_xlabel(r'$x_%d$' % (comb[0]+1), fontsize=FONTSIZE)
    axs[1,i].set_xlabel(r'$\tilde{x}_%d$' % (comb[0]+1), fontsize=FONTSIZE)
    # Set ylabels
    axs[0,i].set_ylabel(r'$x_%d$' % (comb[1]+1), fontsize=FONTSIZE)
    axs[1,i].set_ylabel(r'$\tilde{x}_%d$' % (comb[1]+1), fontsize=FONTSIZE)
    if np.abs(x_arr_t[comb[0],:].max() - x_arr_t[comb[0],:].min()) < 0.0001:
        axs[1,i].set_xlim(left=-1,right=1)
    if np.abs(x_arr_t[comb[1],:].max() - x_arr_t[comb[1],:].min()) < 0.0001:
        axs[1,i].set_ylim(bottom=-1,top=1)
    for ax in [axs[0,i], axs[1,i]]:
        ax.tick_params(labelsize=FONTSIZE-2)
plt.tight_layout()
```
