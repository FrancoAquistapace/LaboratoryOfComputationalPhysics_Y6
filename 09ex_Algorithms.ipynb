1\. **Maximum wind speed prediction at the Sprogø station**

The exercise goal is to predict the maximum wind speed occurring every 50 years even if no measure exists for such a period. The available data are only measured over 21 years at the Sprogø meteorological station located in Denmark. 

The annual maxima are supposed to fit a normal probability density function. However such function is not going to be estimated because it gives a probability from a wind speed maxima. Finding the maximum wind speed occurring every 50 years requires the opposite approach, the result needs to be found from a defined probability. That is the quantile function role and the exercise goal will be to find it. In the current model, it is supposed that the maximum wind speed occurring every 50 years is defined as the upper 2% quantile.

By definition, the quantile function is the inverse of the cumulative distribution function. The latter describes the probability distribution of an annual maxima. In the exercise, the cumulative probability $p_i$ for a given year i is defined as $p_i = i/(N+1)$ with $N = 21$, the number of measured years. Thus it will be possible to calculate the cumulative probability of every measured wind speed maxima. From those experimental points, the scipy.interpolate module will be very useful for fitting the quantile function. Finally the 50 years maxima is going to be evaluated from the cumulative probability of the 2% quantile.

Practically, load the dataset:

```python
import numpy as np
max_speeds = np.load('max-speeds.npy')
years_nb = max_speeds.shape[0]
```

Compute then the cumulative probability $p_i$ (`cprob`) and sort the maximum speeds from the data. Use then the  UnivariateSpline from scipy.interpolate to define a quantile function and thus estimate the probabilities.

In the current model, the maximum wind speed occurring every 50 years is defined as the upper 2% quantile. As a result, the cumulative probability value will be:

```python
fifty_prob = 1. - 0.02
```

So the storm wind speed occurring every 50 years can be guessed as:

``` python
fifty_wind = quantile_func(fifty_prob)
```



````python
# Import modules and load data
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

max_speeds = np.load('max-speeds.npy')
years_nb = max_speeds.shape[0]
````

````python
# Calculate cumulative probabilities and quantile func
cprob = (np.arange(years_nb) + 1) / (years_nb+1) 
quantile_func = UnivariateSpline(x=cprob, y=np.sort(max_speeds))

# Define probability for fifty year period
fifty_prob = 1. - 0.02

# Get estimation of max wind speed in fifty years
fifty_wind = quantile_func(fifty_prob)
print('Maximum estimated wind speed in a 50-year period:',
      np.round(fifty_wind, 2))
````

````python
# Plot the quantile so that we understand the prediction
dom = np.linspace(0,1., num=100)
y_vals = quantile_func(dom)
plt.plot(dom, y_vals)
plt.vlines(x=fifty_prob, 
           ymin=y_vals.min(), ymax=y_vals.max(), 
           color='r', linestyles='--')
plt.ylim(top=y_vals.max(), bottom=y_vals.min())
plt.ylabel('Wind speed')
plt.xlabel(r'$P$');
````

2\. **Curve fitting of temperature in Alaska** 

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18

min: -62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58

* Plot these temperature extremes.
* Define a function that can describe min and max temperatures. 
* Fit this function to the data with scipy.optimize.curve_fit().
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

````python
# First let's store and plot the temperature extremes
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 
         'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temp_max = [17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18]
temp_min = [-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, 
            -52, -58]

# Now the plot
FONTSIZE = 20
fig, ax = plt.subplots(figsize=(8,5), nrows=1, ncols=1,
                      dpi=200)
plt.scatter(months, temp_max, label='Max')
plt.scatter(months, temp_min, label='Min')
plt.xlabel(r'Month', fontsize=FONTSIZE)
plt.ylabel(r'$T$ (°C)', fontsize=FONTSIZE)
plt.legend(fontsize=FONTSIZE-2)
plt.tick_params(labelsize=FONTSIZE-2)
plt.tight_layout();
````

````python
# Import curvefit
from scipy.optimize import curve_fit
````

````python
# Given this is a cyclic pattern, we'll
# propose a sin(x) model

# Let's get the numeric domain of the model
month_nums = np.arange(12)

# Define model
def sin_temp(t, phi, A, B):
    '''
    Params:
        t : int, float or numpy array
            Independent temporal variable, must
            correspond to the number of a month - 1.
            Thus, in the range [0, 11].
        phi : int or float
            Phase of the model.
        A : int or float
            Scale parameter of the model.
    Output:
        Returns y defined as follows:
            y = A * np.sin((t + phi) * (pi/6)) + B
    '''
    return A * np.sin(t * (np.pi/6) + phi) + B


# Now let's fit the model to each dataset
model_temp_max = curve_fit(sin_temp, month_nums, temp_max, 
                           p0=[1,1,np.mean(temp_max)])
model_temp_min = curve_fit(sin_temp, month_nums, temp_min, 
                           p0=[1,1,np.mean(temp_min)])

# Get predictions
dom = np.linspace(0,11, num=100)
preds_temp_max = sin_temp(dom, *model_temp_max[0])
preds_temp_min = sin_temp(dom, *model_temp_min[0])

# Plot the results
FONTSIZE = 20
fig, ax = plt.subplots(figsize=(8,5), nrows=1, ncols=1,
                      dpi=200)
plt.scatter(month_nums, temp_max, label='Max', c='tab:orange')
plt.scatter(month_nums, temp_min, label='Min', c='tab:red')
plt.plot(dom, preds_temp_max, label='Model', c='tab:blue', ls='--')
plt.plot(dom, preds_temp_min, c='tab:blue', ls='--')
plt.xlabel(r'Month', fontsize=FONTSIZE)
plt.ylabel(r'$T$ (°C)', fontsize=FONTSIZE)
plt.legend(fontsize=FONTSIZE-2)
plt.xticks(month_nums, labels=months)
plt.tick_params(labelsize=FONTSIZE-2)
plt.tight_layout();
````

The result looks very reasonable considering the simplicity of the model.

````python
print('Time offset between max and min models: %.1f months' %\
                      (((model_temp_max[0][0]+12)%12)-\
                       ((model_temp_min[0][0]+12)%12)))

perc_error_temp_max = 100*np.mean(np.abs((
    sin_temp(month_nums, *model_temp_max[0]) - temp_max)/np.array(temp_max)))
perc_error_temp_min = 100*np.mean(np.abs((
    sin_temp(month_nums, *model_temp_min[0]) - temp_min)/np.array(temp_min)))
print('\nPercentual absolute mean error:')
print('Max temp. model: %.2f%s' % (perc_error_temp_max, '%'))
print('Min temp. model: %.2f%s' % (perc_error_temp_min, '%'))
````
